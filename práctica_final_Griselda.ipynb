```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca e nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]
columns = ["alumno", "apellido", "course", "marks"]
```

```python
df = pd.DataFrame(data=alumnos, columns=columns)
df.head(14)
```

```python
columns = ["alumno", "apellido", "course", "marks"]
df = pd.DataFrame(data=alumnos, columns=columns)
df.head()
marks = df["marks"].tolist()
print(marks)

course = df["course"].tolist()
print(course)
average_final = np.average(marks, axis=1)
print(average_final)
```

```python
alumno = list(df["alumno"])
apellido = list(df["apellido"])
course = list(df["course"].astype(int))
average_final = list(np.average(marks, axis=1))
ficha_alumno = lambda persona: {'alumno': persona[0], 'apellido': persona[1], 'course': persona[2], 'average_final': persona[3]}
ficha_alumno_final = list(map(ficha_alumno,zip(alumno, apellido, course, average_final)))
print(ficha_alumno_final)
```

```python
df3 = pd.DataFrame.from_dict(ficha_alumno_final)
df3
```
